```python
from google.cloud import aiplatform
import time

# Endpoint 객체 생성
endpoint = aiplatform.Endpoint(
    endpoint_name="projects/953105731059/locations/us-central1/endpoints/6923600530849660928"
)

# Payload 설정
payload = {
  "instances": [
    {
        "body": "Oh okay I'm gonna try it, No. I dont like abocado."
    }
  ]
}

# API 호출 전 시간 기록
start_time = time.time()

# API 호출
response = endpoint.predict(instances=payload['instances'])

# API 호출 후 시간 기록
end_time = time.time()

# 호출 속도 계산
duration = end_time - start_time

# 출력
print(f"API call duration: {duration:.2f} seconds")
```
